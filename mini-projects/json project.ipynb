Now I switch to looking for the top 10 major project themes

```python
df.mjtheme_namecode.value_counts().head(10)
```

It is apparent that this needs to be loaded as a string to allow for normalizing of the data

```python
df_string = json.load((open('data/world_bank_projects.json')))
project_themes = json_normalize(df_string, 'mjtheme_namecode')
project_themes
```

Again, looking for the top 10 themes

```python
project_themes.name.value_counts().head(10)
```

It is noted that the number 7 spot is, in fact, blank names. Next, it was checked to see if the code number was complete

```python
project_themes.code.value_counts()
```

There are no blank code numbers, so the names can be deduced from their code number. First, the table was sorted by code and name

```python
themes_sorted = project_themes.sort_values(['code', 'name'])
themes_sorted
```

Then, the blank spaces were filled with 'NaN' to be replaced later

```python
themes_sorted.name[themes_sorted['name'] == ''] = np.nan
themes_sorted
```

Now, the 'NaN' spaces can be backfilled with the correct name

```python
themes_sorted = themes_sorted.fillna(method='bfill')
themes_sorted
```

Again, printed the top 10 project themes, now that all blank spaces are filled

```python
themes_sorted.name.value_counts().head(10)
```

```python
new_df = themes_sorted.sort_values(['code','name'])
new_df
```
